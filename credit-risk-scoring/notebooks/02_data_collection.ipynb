```python
from sklearn.datasets import fetch_openml
import pandas as pd
```

```python
data = fetch_openml('credit-g', as_frame=True)

df = data.frame

print(df.head())
print(df.columns)
```

```python
df.rename(columns={'class': 'Risk'}, inplace=True)

# convert to binary
df['Risk'] = df['Risk'].map({'good': 0, 'bad': 1})
```

```python
import os

os.makedirs('data/raw', exist_ok=True)
```

```python
df.to_csv('data/raw/german_credit.csv', index=False)
```
